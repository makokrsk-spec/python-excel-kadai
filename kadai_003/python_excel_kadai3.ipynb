```python
import pandas as pd

# データフレームを作る
df = pd.DataFrame()

# ファイル名のリストを格納する
file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

# ファイルを一つずつ処理する
for file in file_list:
    # ファイルを読み込む
    df_temp = pd.read_excel(file)

    # データフレームの列名を指定する
    df_temp.columns = ['売上年', '商品', '金額（千円）']

    # concat()でデータフレームを連結する
    df = pd.concat([df, df_temp], ignore_index=True)

# 「商品」「売上年」ごとにグループ化し、合計金額を算出する
df = df.groupby(['商品', '売上年'])['金額（千円）'].sum()

df_excel = df.reset_index()

# Excelファイルを作成する
writer = pd.ExcelWriter('売上集計表.xlsx')

# DataFrameオブジェクトをExcelファイルに書き込む
df_excel.to_excel(writer, sheet_name='Sheet1', index=False)

# Excelファイルを保存する
writer.close()

#print("売上集計表.xlsx が作成されました。")

import openpyxl
# openpyxl.stylesモジュールのPatternFillクラスをインポート
from openpyxl.styles import PatternFill

# Excelファイルを読み込む
wb = openpyxl.load_workbook('売上集計表.xlsx')
ws = wb.active

# 色を指定する
header_fill = PatternFill(patternType='solid', fgColor='F2F2F2')

# ヘッダー部分（A1からC1）を薄いグレー（#F2F2F2）に設定する
for row in ws['A1:C1']:
    for cell in row:
      cell.fill = header_fill

# Excelファイルを保存する
wb.save('売上集計表.xlsx')
```
