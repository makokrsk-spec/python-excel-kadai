```python
import openpyxl
from datetime import datetime

# 新しいワークシートを開く
workbook = openpyxl.Workbook()
ws = workbook.active

# 固定データを書き込む
ws['B2'] = '請求書'
ws['B4'] = '株式会社ABC'
ws['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
ws['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B7'] = '担当者名:鈴木一郎 様'

# 日付を設定する
date_day = datetime.now()

# No.と日付を書き込む
ws['F4'] = 'No.'
ws['G4'] = '001'
ws['F5'] = '日付'
ws['G5'] = date_day.strftime('%Y/%m/%d')

# 表のヘッダーを設定する
header = ['商品名', '数量', '単価', '金額']

# 表にデータを書き込む
data = [
    ['商品A', 2, 10000],
    ['商品B', 1, 15000]
]

# 表を書き出す場所を指定する
ws.cell(row=10, column=2, value=header[0]) # ヘッダー
ws.cell(row=10, column=3, value=header[1])
ws.cell(row=10, column=4, value=header[2])
ws.cell(row=10, column=5, value=header[3])

start_row = 11

for i, datum in enumerate(data):
    ws.cell(row=start_row+i, column=2, value=datum[0]) # 商品名、数量、単価
    ws.cell(row=start_row+i, column=3, value=datum[1])
    ws.cell(row=start_row+i, column=4, value=datum[2])

    amount = datum[1] * datum[2]
    ws.cell(row=start_row+i, column=5, value=amount) # 金額

# 小計を書き込む行を追加して金額を書き込む
last_data_row = ws.max_row
total_row = last_data_row +1
total_value = 0
for i, datumn in enumerate(data):
    total_value = total_value + datumn[1] * datumn[2]
    ws.cell(row=total_row, column=5, value=total_value) # 小計

# 小計、消費税、税込合計を計算して書き込む
tax_rate = 0.1
tax = total_value * tax_rate
total_all = total_value + tax

ws.cell(row=total_row+2, column=2, value='小計')
ws.cell(row=total_row+2, column=5, value=total_value)
ws.cell(row=total_row+3, column=2, value='消費税')
ws.cell(row=total_row+3, column=5, value=tax)
ws.cell(row=total_row+4, column=2, value='合計')
ws.cell(row=total_row+4, column=5, value=total_all)

today = datetime.now()
workbook.save(f'請求書_{today.strftime('%Y%m%d')}.xlsx')
```
